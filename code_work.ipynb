2) Write a code to inter-convert the 3 graph representations.

```python
#  adjaceny list to edge list :

def adj_edge(dic):
    edge=[] # empty edge list to store edges as tupple
    for node in dic: # loop through each node in adjaceny list
        for adjacent_node in dic[node]: # iterate its adjacent nodes
            if (adjacent_node, node) not in edge: # avoid duplicates
                edge.append((node, adjacent_node)) # add the edge to the edge list 
    return edge # return the complete edge list 

#  edge lsit => adjaceny list 
def edge_dic(edge):
    dic = {}  # Initialize empty dictionary
    for first, second in edge:  # Loop through each edge (pair of nodes)
        if first not in dic:  # If node not in dictionary, add it
            dic[first] = []
        if second not in dic:
            dic[second] = []
        dic[first].append(second)  # Add connection
        dic[second].append(first)  # Add reverse connection 
    return dic  # Return adjacency list

#  Matrix => Adjacency List
def matrix_dic(matrix, name):
    dic = {}  # Initialize empty dictionary
    for i in range(len(name)):  # Loop through each node index
        dic[name[i]] = []  # empty list for that node
        for j in range(len(name)):  # Check all possible connections
            if matrix[i][j] == 1:  # If there's an edge
                dic[name[i]].append(name[j])  # Add connection
    return dic  # Return adjacency list

# matrix => edge list 
def matrix_edge(matrix, name):
    edge = []  # Empty edge list
    for i in range(len(name)):  # Loop through rows
        for j in range(len(name)):  # Loop through columns
            if matrix[i][j] == 1:  # If there's a connection
                # Avoid duplicate edge
                if (name[j], name[i]) not in edge:
                    edge.append((name[i], name[j]))  # Add edge
    return edge  # Return edge list

# edge lsit => matrix 
def edge_matrix(edge, name):
    size = len(name)  # Number of nodes
    matrix = []  # Final matrix

    # Step 1:  x size matrix filled with 0s
    for i in range(size):
        row = []
        for i in range(size):
            row.append(0)
        matrix.append(row)

    # Step 2: Loop through each edge and mark 1 in the matrix
    for pair in edge:
        first_name = pair[0]
        second_name = pair[1]

        # Find index of both nodes
        first_index = -1
        second_index = -1
        for i in range(size):
            if name[i] == first_name:
                first_index = i
            if name[i] == second_name:
                second_index = i

        # make 1 in both way is that make a connection both ways 
        if first_index != -1 and second_index != -1:
            matrix[first_index][second_index] = 1
            matrix[second_index][first_index] = 1

    return matrix  # Return adjacency matrix


#  adjaceny => matrix
def list_to_matrix(dic, name):
    size = len(name)  # Number of nodes
    matrix = []  # Final matrix

    # Step 1: Create empty matrix with all 0s
    for i in range(size):
        row = []
        for i in range(size):
            row.append(0)
        matrix.append(row)

    # Step 2: Mark 1 where nodes are connected
    for i in range(size):
        node = name[i]
        for connected in dic[node]:
            # Find index of the connected node
            index_of_connected = -1
            for j in range(size):
                if name[j] == connected:
                    index_of_connected = j
                    break
            # Mark the connection in matrix
            if index_of_connected != -1:
                matrix[i][index_of_connected] = 1

    return matrix  # Return adjacency matrix

```

```python
#  example 
# adjacency list
dic = {
    's1': ['s2', 's3'],
    's2': ['s1'],
    's3': ['s1']
}

# edge list
edge = [('s1', 's2'), ('s1', 's3')]

# matrix
matrix = [
    [0, 1, 1],  # s1 is connected to s2 and s3
    [1, 0, 0],  # s2 is connected to s1
    [1, 0, 0]   # s3 is connected to s1
]

# name lsit of nodes 
name = ['s1', 's2', 's3']

print("List => Edge:")
print(adj_edge(dic))

print("\nEdge => List:")
print(edge_dic(edge))

print("\nMatrix => List:")
print(matrix_dic(matrix, name))

print("\nMatrix => Edge:")
print(matrix_edge(matrix, name))

print("\nEdge => Matrix:")
mat1 = edge_matrix(edge, name)
for row in mat1:
    print(row)

print("\nList => Matrix:")
mat2 = list_to_matrix(dic, name)
for row in mat2:
    print(row)

```

4) check if a graph is complete

```python
#  to check if a graph is complete based on  adjacency matrix
def complete(graph):
    # Loop through each row index(i) of the matrix
    for i in range(len(graph)):

        # Loop through each column index(j) of the matrix
        for j in range(len(graph)):

            #  Case 1: i == j means diagonal
            if i == j:
                # in undirected graph there is no self loop
                # So the diagonal should be 0
                if graph[i][j] != 0:
                    return 'self connected'  #  not complete graph

            #  Case 2: i != j  check if there is connection between each node other than diagonal 
            else:
                # In a complete graph, every distinct pair of nodes must be connected
                # So graph[i][j] should be 1
                if graph[i][j] != 1:
                    return 'incomplete'  # if there exists two nodes which dies not have any nodge 

    # if all conditions satisfied return complete graph
    return 'complete'
# Example of a complete graph (4 nodes)
matrix = [
    [0, 1, 1, 1],
    [1, 0, 1, 1],
    [1, 1, 0, 1],
    [1, 1, 1, 0]
]

print(complete(matrix))  # Output: 'complete'
```

 5) check if a graph is connected

```python
# check if  agraph is connected using queue method 
def connect(graph, node_list ): # node_list which store every node
    visited = []  # list to store visited nodes
    queue = []    # to iterate through nodes 
    
    # Step 1: Start from the first node in the node_list
    start = node_list[0]
    visited.append(start) # add it to the visited lsit
    queue.append(start) # Add to the queue to iterate its adjacent nodes

        # Step 2: dequeue nodes in the queue one by one 
    while queue:
            n = queue.pop(0)  # Remove the front node from the queue 
            if n in graph:
                # Traverse neighbors of the node 'n'
                for neighbor in graph[n]:  # Iterate over the neighbors of 'n'
                    if neighbor not in visited:  # If the neighbor is unvisited
                        visited.append(neighbor)  # Mark it as visited
                        queue.append(neighbor)    # Add it to the queue for further processing
    
    # Step 3: traverse whole grpah and check if we visited all nodes or not
    # name list is a list of names of all nodes 
    if len(visited) == len(node_list):
        return ('connected graph', visited)  #  Graph is connected
    else:
        return 'disconnected'  #  Graph is disconnected 
dic = {
    's1': ['S2'],
    'S2': ['s1'],
    's3': [],   # Node 'C' is disconnected
    's4': []
}
name = ['s1', 'S2', 's3', 's4']

print(connect(dic, name)) 
```

6) Given a graph and a list of vertices, check if it forms a walk, or a trail or a path or none of these.

```python
# check if route is a walk
def is_walk(graph, route):
    # Step 1: Go through vertices pair 
    for i in range(len(route)-1):
        # Step 2: Check if the next vertice in the route exist in the adjacency list of the present iterating node
        if route[i+1] not in graph[route[i]]:
            return False  # If not, it's not a walk
    return True  # If all are connected, it's a walk


# to check if it is a trail

# to check if a given route  is a trail
def is_trail(graph, route):
    # A trail must have at least two nodes to form at least one edge
    if len(route) < 2:
        # If route has less than 2 nodes, it cannot have any edge
        return False

    # Initialize an empty list to keep track of vsiited edges
    # This helps ensure no edge is repeated 
    vsiited_edges = []

    # Loop through each consecutive pair of nodes in the path
    # check the edge between route[i] and route[i+1]
    for i in range(len(route) - 1):
        #  first node of the iterating edge
        node1 = route[i]
        #  second node of the iterating edge
        node2 = route[i + 1]

        # check if node1 exists in the graph
        # If not, the route is invalid since that node doesn't exist
        if node1 not in graph:
            return False

        #  check if node2 is connected to node1 if its present in its adjacency list 
        connected = False  # Assume they are not connected

        # Loop through all adjacent nodes  of node1 to find node2
        for i in graph[node1]:
            # If node2 is found in node1's adjacency list
            if i == node2:
                connected = True  # They are connected
                break  # No need to check further

        # If node2 is not found node1's neighbors, the edge doesn't exist
        if not connected:
            return False  # nto a trail as not an edge between them

        # Create an edge tuple from node1 to node2
        edge = (node1, node2)

        # check if that edge is been visited or not
        # in undirected graph also check for reversed direction, as both edges are same 
        for used_edge in vsiited_edges:
            # If ewdge(node1, node2) or (node2, node1) is in visited_edge
            if used_edge == (node1, node2) or used_edge == (node2, node1):
                return False  # edges are not repeated in trail

        # If edge not in the visited list append it
        vsiited_edges.append(edge)

    # If all edges are valid and no edge was repeated
    # tehn the route is a trail
    return True

# check if the route is a path
def is_path(route):
    visited = []  # To store visited vertices

    # Step 1: traverse through all vertices in the route
    for i in range(len(route)):
        repeat = False

        # Step 2: Check if current vertex already visited
        for j in range(len(visited)):
            if visited[j] == route[i]:
                repeat = True
                break

        if repeat == True:
            return False  # Repeated vertex/node means it's not a path

        visited.append(route[i])  # Add to visited list

    return True  # If all vertices are unique, it's a path

# check  if its walk, trail, path or none of these :
def check_type(graph, route):
    if is_walk(graph, route) == False:
        return 'Not a walk'
    elif is_trail(graph, route) == False:
        return 'Walk'  # It's a walk but not a trail
    elif is_path(route) == False:
        return 'Trail'  # It's a trail but not a path
    else:
        return 'Path'  # It's a walk + trail + path


# Example graph (adjacency list form)
graph = {
    's1': ['s2', 's3'],
    's2': ['s1', 's4'],
    's3': ['s1', 's4'],
    's4': ['s2', 's3']
}

print(check_type(graph, ['s1', 's2', 's4', 's3']))  
print(check_type(graph, ['s1', 's3', 's4', 's2','s1']))  
print(check_type(graph, ['s1', 's2', 's1', 's3']))  
print(check_type(graph, ['s1', 's4', 's2']))        
```

7) Check if a given graph is a tree.

 => conditions for a graph to be a tree:
 must be connected
 should not have any cycle(acyclic)

 =>graph with n nodes and n-1 edges ensure it .

```python
#  to check if the graph has a cycle

def check_cycle(graph, node, visited, parent):
    visited.append(node)  # add the current node in visited
    
    # Traverse through all adjacent nodes of the current node
    for adj in graph[node]:
        
        # If the adjacent node has not been visited
        if adj not in visited:
            # Recursively check for cycles in the subtree
            if check_cycle(graph, adj, visited, node):
                return True  # If a cycle is found in any subtree, return True

        # If the adjacent node is already visited and is not the parent
        # which menas we came back to previous node (cycle found)
        elif adj != parent:
            return True  # Found a cycle

    return False  # No cycle found from this node


# Function to check if a graph is a tree
def is_tree(graph):
    # Step 1: Use connect() function to check if graph is connected
    all_nodes = list(graph.keys())         # Get the list of all node names
    result = connect(graph, all_nodes)     # call connect fucntion to check if grpah is connected

    # Step 2: If the graph is not connected, it can't be a tree
    if result == 'disconnected':
        return False

    # Step 3: Check if the graph has a cycle
    start = all_nodes[0]           # Choose first node 
    visited = []                   # store all visitied nodes

    # If has a cycle, it's not a tree
    if check_cycle(graph, start, visited, None):
        return False

    # Step 4: If connected and no cycles, then it's a tree
    return True


#  Not a tree (contains a cycle: s1 → s2 → s3 → s1)
graph1 = {
    's1': ['s2', 's3'],
    's2': ['s1', 's3'],
    's3': ['s1', 's2'],
    's4': []  # disconnected node
}
print("Graph 1 is a tree:", is_tree(graph1))  


#  This is a tree (connected and has no cycles)
graph2 = {
    's1': ['s2', 's3'],
    's2': ['s1', 's4'],
    's3': ['s1'],
    's4': ['s2']
}
print("Graph 2 is a tree:", is_tree(graph2))  
```

8) Given a connected cyclic graph, find its spanning tree.

=> remove edges such that the graph still remains connected
=> no cycles


```python
import random  # import random to use random function 

#  to create a spanning tree 
def spanning_tree(graph):
    # STEP 1: Pick a random node to sart with 
    node_list = list(graph.keys())           # list of keys/ nodes 
    start_node = random.choice(node_list)    # randomly choose a node to start with
    print("Random  Node:", start_node)
    # STEP 2: use queue mehtod to generate the spanning tree
    visited = []        # to store the visited nodes
    queue = []          #queue to store the next node to visit
    tree = []     # to store the final tree edges.

    # visit the first node we randomly chose .
    visited.append(start_node)
    queue.append(start_node)

    # Continue until the queue is empty
    while queue:
        n = queue[0]       # to extract first node from the queue
        queue = queue[1:]        # Remove it from the queue

        # check adjacent nodes of the n
        for adj in graph[n]:
            if adj not in visited:
                visited.append(adj)                 # append adjacent node to visited
                tree.append((n, adj))   # Add edge to tree
                queue.append(adj)                   # Add adjacent node to queue

    # return the final spanning tree
    return tree


# Example:
graph_data = {
    'S1': ['S2', 'S3', 'S4'],
    'S2': ['S1', 'S4'],
    'S3': ['S1', 'S4'],
    'S4': ['S1', 'S2', 'S3']
}

# Call the function
spanning_tree = spanning_tree(graph_data)

# Print the result
print("spanning tre")
for edge in spanning_tree:
    print(edge)
```

11) Given a tree find its height 
=> if leaf node height 0
=> else height is 1 + max height of its children nodes

```python
# to calculate the height of the tree:

def height(node, tree, parent=None, visited=None): # in the starting parent is none as the height for root node 
    #  visited is none as none of the node is visited 
    # initialise a visited node to ensure the child node does not have parent node as its child
    if visited is None:
        visited=[]
    visited.append(node) # add the node to visited list 
    # Check if the node is a leaf 
    if len(tree[node]) == 1 and parent in tree[node]:
        return 0
    
    # caluclate the height for all children
    ht_list = []
    for child in tree[node]:
        if child != parent and child not in visited :  # Skip the parent node to avoid cycling and also already visitted node 
            ht_list.append(height(child, tree, node, visited))  # now make the current node as parent and repeat the process
    
    # Simplify the max logic: if heights is empty, return 0, else return max height + 1
    
    if ht_list:
        return 1 + max(ht_list) # height of the parent is 1+ child nodes height 
    else:
        return 0 # if the lsit is empty

#tree's height assuming root node is s1
tree = {
    'S1': ['S2', 'S3', 'S4'],
    'S2': ['S1', 'S4'],
    'S3': ['S1', 'S4'],
    'S4': ['S1', 'S2', 'S3']
}
print("Height of  tree:", height('S1', tree))

```

12) given a tree find its depth==> length of path from root node to the deepest leaf node 

<!--  height of a a tree is saame as depth of a tree numerically -->
for particular node it varies, but for the tree it is same 

```python
# Function to calculate the depth of the tree
def depth(node, tree, parent=None, visited=None):  # parent is None initially as we start from root
    if visited is None:
        visited = []  # initialize visited list

    visited.append(node)  # mark the iterated node as visited

    # Check if the current node is a leaf node
    if len(tree[node]) == 1 and parent in tree[node]:
        return 0  # leaf node has depth 0

    # calculate depth for all child nodes
    dp_list = [] # store depth of child nodes
    for child in tree[node]:
        if child != parent and child not in visited:  # skip the parent and already visited nodes
            dp_list.append(depth(child, tree, node, visited))  # make current node as parent and repeat

    #  if list is empty return 0 else return 1 + max of children
    if dp_list:
        return 1 + max(dp_list)  # depth is 1 + maximum depth of children
    else:
        return 0  # if no child exists

# Example 
tree = {
    'S1': ['S2', 'S3', 'S4'],
    'S2': ['S1', 'S4'],
    'S3': ['S1', 'S4'],
    'S4': ['S1', 'S2', 'S3']
}

# Assuming root node is S1
print("Depth of tree:", depth('S1', tree))

```
